# Learning the ADNI Dataset

#### Reading the data:

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


data = pd.read_csv("DATA/ADNI_dataset.csv", low_memory=False) # Need the low_memory or dtypes warning
data.head(10)
```

#### Some columns that interest us:

```python
columns_of_interest = ['PTID',
                       'Date',
                       'Age', 
                       'Sex',
                       'APOE4_Alleles',
                       'APOE_Genotype', 
                       'Family_History_Dementia', 
                       'Diabetes', 
                       'Hypertension',
                       'Alcoholic',
                       'Cholesterol',
                       'Diastole', 
                       'Systole', 
                       'Smoking',
                       'BMI',
                       'Diagnosis']
c = list(data.columns)
MRI_columns = c[c.index("MUSE_Volume_4"):c.index("MUSE_Volume_207")+1]

columns_of_interest += MRI_columns

data_of_interest = data[columns_of_interest]
data_of_interest.head(10)
```

#### Find the unique participants:

```python
unique = data_of_interest.drop_duplicates(subset='PTID')
print(unique.shape)
unique.head(10)
```

#### Visualizations:

```python
print(unique['Sex'].value_counts())
unique['Sex'].value_counts().plot(kind='bar', figsize=(8,8), rot=0);
plt.xlabel("Gender", labelpad=14);
plt.ylabel("Count of People", labelpad=14);
```

```python
print(unique.groupby('Sex').Age.mean())
unique.groupby('Sex').Age.hist(legend=True, bins=30, alpha=0.5, figsize=(8,8));
```

```python
print(unique['APOE_Genotype'].value_counts())
print(unique['APOE4_Alleles'].value_counts())
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(15,9));
unique.APOE_Genotype.value_counts().plot(kind='bar', ax = axes[0], subplots=True, rot=0);
unique.APOE4_Alleles.value_counts().plot(kind='bar', ax = axes[1], subplots=True, rot=0);
```

```python
print(unique['Diabetes'].value_counts()); print()
print(unique['Hypertension'].value_counts()); print()
print(unique['Alcoholic'].value_counts()); print()
print(unique['Smoking'].value_counts()); print()
print(unique['Family_History_Dementia'].value_counts()); print()
print(unique['Diagnosis'].value_counts()); print()


fig, axes = plt.subplots(nrows=2,ncols=3,figsize=(15,15));
unique.Diabetes.value_counts().plot(kind='bar', ax = axes[0,0], subplots=True, rot=5);
unique.Hypertension.value_counts().plot(kind='bar', ax = axes[0,1], subplots=True, rot=5);
unique.Alcoholic.value_counts().plot(kind='bar', ax = axes[0,2], subplots=True, rot=0);
unique.Smoking.value_counts().plot(kind='bar', ax = axes[1,0], subplots=True, rot=0);
unique.Family_History_Dementia.value_counts().plot(kind='bar', ax = axes[1,1], subplots=True, rot=0);
unique.Diagnosis.value_counts().plot(kind='bar', ax = axes[1,2], subplots=True, rot=0);
```
